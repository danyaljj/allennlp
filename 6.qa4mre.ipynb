```python
from prereq import Prerequisit_Skills

preq = Prerequisit_Skills()
# print(preq.skills_map)
prerequisits = preq.read_prereq_file("qa4mre.json")

# ignore the nonsense content 
prerequisits = {x: prerequisits[x]  for x in prerequisits.keys() if 'Nonsense' not in prerequisits[x] }
# print(len(prerequisits))
# print(set(prerequisits.keys()))
# print(len(set(prerequisits.keys())))
prerequisits
```

```python
# convert the questions into the squad format 
paragraphs = []
for topic in data: 
    topic_id = topic["topic_id"]
    topic_name = topic["topic_name"]
    for reading in topic["reading-test"]: 
        reading_id = reading["reading_id"]
        document_text = reading["doc"]["#text"]
        qas = []
        for q in reading["q"]: 
            question_id = q["question_id"]
            q_text = q["q_str"]
            ans1 = []
            for a in q["answer"]: 
                answer_id = a["answer_id"]
                answer_text = a["#text"]
                idx = 0
                if answer_text in document_text:
                    idx = document_text.index(answer_text)
                ans1.append({"answer_start": idx, "text": answer_text})
            question = {"answers": ans1, "question": q_text.strip(), "id": topic_name + "-" + reading_id + "-" + str(question_id)}
            qas.append(question)
        paragraphs.append({"context": document_text.strip(), "qas": qas})
        
questions_json = {"data": [{"paragraphs": paragraphs}]}
with open('/Users/daniel/ideaProjects/allennlp_invariances/QA_datasets/QA4MRE/QA4MRE-2013-EN_test_squad_format.json', 'w', newline='') as f:
    f.write(json.dumps(questions_json))
```
