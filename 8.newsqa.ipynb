```python
file = "/Users/daniel/ideaProjects/allennlp_invariances/QA_datasets/newsqa/combined-newsqa-data-v1.json"
import json

with open(file) as f:
    data = json.load(f)
```

```python
# for x in data
# print(len(data["data"]))
# print(data["data"][0]["questions"])
paragraphs = []
for x in data["data"]: 
    text = x["text"].strip()
    storyId = x["storyId"]
    questions = x["questions"]
    qas = []
    for qId, q in enumerate(questions): 
        ans1 = []
        question_text = start_index = q["q"].strip()
        if "s" in q["consensus"]: 
            start_index = q["consensus"]["s"]
            end_index = q["consensus"]["e"]
            answer_string = text[start_index:end_index].strip()
            ans1.append({"answer_start": start_index, "text": answer_string})
            question = {"answers": ans1, "question": question_text, "id": storyId + "-" + str(qId)}
            qas.append(question)
    paragraphs.append({"context": text, "qas": qas})
questions_json = {"data": [{"paragraphs": paragraphs}]}
with open("/Users/daniel/ideaProjects/allennlp_invariances/QA_datasets/newsqa/combined-newsqa-data-v1-squad-format.json", 'w', newline='') as f:
    f.write(json.dumps(questions_json))
```
